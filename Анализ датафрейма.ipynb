Подключаем необходимые библиотеки, загружаем данные и создаём датафрейм

```python
!pip install bootstrapped
from google.colab import drive
import pandas as pd
drive.mount
data = pd.read_csv('/content/sample_data/fifa_s2.csv')

data.head(10)
```

Выполним команду по удалению столбцов, содержащих одинаковые значения

```python
data = data.T.drop_duplicates().T
data.head(10)
```

Столбцов, содержащих одинаковые значения не обнаружено

```python
data.info()
```

Статистика по незаполненным полям

```python
data.isnull().sum()
```

Преобразуем названия клубов к нижнему регистру, во избежания двоякости прочтения

```python
data['Club'] = data['Club'].str.lower()
data.head(10)
```

```python
data.drop_duplicates(subset=['ID'])
data.info()
```

Удалим во всех столбцах пробелы в начале и в конце строки, дабы избежать двоякости прочтения

```python
data = data.applymap(lambda x: x.strip() if isinstance(x, str) else x)
data.head(10)
```

Удалим дубликаты по столбцам имя, возраст, национальность и клуб

```python
data.info()
data = data.drop_duplicates(subset=['Name', 'Age', 'Nationality', 'Club'])
data.info()
```

были удалены 4 строки, как дублирующие

Создадим группы возрастов, определив их границы, затем классифицируем весь датафрейм по этим условиям

```python
bins = [0, 20, 30, 36, 150]
labels = ['до 20', 'до 30', 'до 36', 'старше 36']
data['Age_group'] = pd.cut(data['Age'], bins=bins, labels=labels, include_lowest=True)
data.head(10)
```

Подсчитаем кол-во футболистов в каждой возрастной группе


```python
age_group_counts = data['Age_group'].value_counts().to_frame('Count')
print(age_group_counts)
```
